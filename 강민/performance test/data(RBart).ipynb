```python
import pandas as pd 

data_row=pd.read_csv("/Users/withmocha/Desktop/DATA/BOAZ/미니 프로젝트 1/data/test data/news.csv",index_col=False)

data_row
```

```python
data_row.columns
```

```python
data_row.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
data_row=data_row.sample(frac=1)
data_row
```

```python
data=pd.DataFrame(index=range(300),columns=['text','label'])
```

```python
for i in range(data.shape[0]):
    data.iloc[i,0]=data_row.iloc[i,0]+data_row.iloc[i,1]
    data.iloc[i,1]=data_row.iloc[i,2]
```

```python
data
```

```python
for i in range(data.shape[0]):
    if data.iloc[i,1]=="REAL":
        data.iloc[i,1]=1.0
    else:
        data.iloc[i,1]=0.0
```

```python
data
```

```python
import torch
from transformers import RobertaTokenizer, RobertaModel

model_name = "roberta-base"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = RobertaModel.from_pretrained(model_name)

texts=data.iloc[3,0]


inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")


with torch.no_grad():
    outputs = model(**inputs)


last_hidden_states = outputs.last_hidden_state
```

```python
A=last_hidden_states.flatten()[:1000]
```

```python
list=[i for i in range(len(A))]
```

```python
vector_data=pd.DataFrame(index=range(data.shape[0]),columns=list)
vector_data
```

```python
for i in range(data.shape[0]):
    texts=data.iloc[i,0]
    print(i,'------------------------------------`-------------------------')
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    last_hidden_states = outputs.last_hidden_state
    vector_data.iloc[i]=last_hidden_states.flatten()[:vector_data.shape[1]]
```

```python
vector_data
```

```python
Y=pd.DataFrame(data['label'],columns=['label'])

data_to_save=pd.concat([vector_data,Y],axis=1)
data_to_save=data_to_save.dropna(axis=0)
data_to_save.to_csv("fake news test vectored data.csv")
```
